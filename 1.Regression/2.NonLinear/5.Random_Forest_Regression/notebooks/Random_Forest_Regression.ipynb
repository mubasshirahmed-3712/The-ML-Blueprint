<h2 style="text-align:center;">Random Forest Regression</h2>

Random Forest Regression is an **ensemble learning technique** that builds multiple decision trees and combines their outputs to improve prediction accuracy and reduce overfitting.  
It is widely used for **non-linear data** where relationships between variables are complex.



### ðŸ“˜ Key Concepts of Random Forest Regression

- **Ensemble Learning**: Combines predictions from multiple models (Decision Trees).  
- **Bagging (Bootstrap Aggregation)**: Each tree is trained on a random subset of data.  
- **Final Prediction**: Average of predictions from all trees.  
- **Reduces Overfitting**: Compared to a single Decision Tree, Random Forest is more stable.

### ðŸ”‘ Important Hyperparameters
- `n_estimators`: Number of trees in the forest.  
- `max_depth`: Maximum depth of each tree.  
- `min_samples_split`: Minimum samples required to split a node.  
- `random_state`: Ensures reproducibility.


```python
# âž¤ Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import joblib

# âž¤ Load dataset
data = pd.read_csv("../data/emp_sal.csv")
X = data.iloc[:, 1:2].values   # Position Level
y = data.iloc[:, 2].values     # Salary

# âž¤ Train Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# âž¤ Predict for a sample value
sample_value = [[4.0]]
predicted_salary = rf_model.predict(sample_value)
print(f"Predicted Salary at Position Level 4.0: {predicted_salary[0]}")  # âž¤ Prediction Output

# âž¤ High-resolution visualization
X_grid = np.arange(min(X), max(X), 0.01)  # finer steps for smooth curve
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color="red", label="Actual Data")
plt.plot(X_grid, rf_model.predict(X_grid), color="blue", label="Random Forest Predictions")
plt.scatter(sample_value, predicted_salary, color="green", s=100, label="Prediction @ 4.0")
plt.title("Random Forest Regression Salary Prediction")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# âž¤ Save trained model to models/ folder
joblib.dump(rf_model, "../model/rf_model.pkl")
print("âœ… Model saved as rf_model.pkl in models/ folder")
```
